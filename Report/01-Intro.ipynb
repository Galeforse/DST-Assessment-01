# Introduction


_This project is an assignment for the Data Science Toolbox module (2020-2021 Academic Year) from the University of Bristol and it has 4 main contributors --See bottom of the page--. The aim of this assessment is consequently **not** to 'win', or get as good of an accuracy on the test results as possible, but rather to present different models and compare them at the end. All contributors have learned new skills and enhanced their knowledge of the subject tackled throughout this work, and as such, there is plenty of room for improvement on both final results and code optimization._

## Project Aim and Stages


The main purpose of this project is to detect and predict abnormal traffic behaviour in [a given dataset](http://kdd.ics.uci.edu/databases/kddcup99/kddcup99.html). Manipulating datapoints and learning from a subset of the whole dataset, called the _training set_ , the contributors were asked to predict what the nature of the traffic will look like on the unused subset of the dataset, which we will refer to as the _testing set_ . The team was allowed to tackle this issue however they wished, with the following guidances:

"Your task for this project is to:

• each create a model submission, with the goal of predicting normal vs
non-normal traffic.

• together agree and test a performance metric.

• compare your models according to that performance metric. "


The expected result is 4 different models, predicting normal v.s non-normal behaviour with better or worse accuracy. The team decided to follow a 3 stage process towards the completion of this project:

**1. Splitting the data**

We have decided to split the data using 10-fold Cross Validation (10CV). This decision comes as the dataset was deemed far too large for LOOCV to yield any appropriate results, and hence k-fold was considered instead. We chose k=10 as it's a very usual choice, and for 'trial-and-error' considerations - such as deciding that k=2,3 would be too small and k>20 too complex for the purposes of this work.

This has resulted in a [training dataset](https://github.com/Galeforse/DST-Assessment-01/blob/main/Data/kd_train.csv.zip), containing 90% of the original dataset, and a [testing dataset](https://github.com/Galeforse/DST-Assessment-01/blob/main/Data/kd_test.csv.zip), containing the remaining 10%. All learningand analysis was done on the former, for the purpose of predicting traffic behaviour on the latter. Exact instructions on how this splitting has been performed are to be found [here](https://github.com/Galeforse/DST-Assessment-01/blob/main/Data/Instructions%20on%20folding.ipynb).

**2. Model Selection**

Each member of the group has chosen a single and unique model for their predictions, fit to the same training and testing datasets. The reasons behind the respective selections are explained withing each contributor's individual folder {Published publicly} and within their independent personal reports {Unpublished publicly}.

This results in 4 independent models, with total freedom of data manipulation on the training set. Each analysis is presented in the [Main Report Folder](https://github.com/Galeforse/DST-Assessment-01/tree/main/Report) - items 02 to 05; as well as found here:

**3. Performance Analysis**

At the end of the project, the contributors come together to compare their models. They will discuss their results; i.e. the accuracy of their predictions on a team picked performance metric. The chosen metric is the Hamming Distance metric, as it fit the binary nature of normal traffic (1) v.s non-normal traffic (0) analysis performed. More details explaining the choice are to be found in the [Main Report Folder](https://github.com/Galeforse/DST-Assessment-01/tree/main/Report) - item 06.

Brief conclusions on why each model gave out their respective performance are drawn, as well as, more importantly, a presentation of the challenges encountered throughout the project and their solutions. All these can be found [here](https://github.com/Galeforse/DST-Assessment-01/blob/main/Report/06-Performance%20Analysis%20and%20Conclusion.ipynb).


**How to interpret the project**

The reader is asked to look at the items chronologically, as presented in the [Report](https://github.com/Galeforse/DST-Assessment-01/tree/main/Report) in order to understand the full extent of this project -- following items 01 up to 06. Alternatively, all the work can be found in the in each member's individual folder; and can be read in whichever order since the models are unrelated to one another. There exists an additional folder in the repository which containg [all the data which has been used](https://github.com/Galeforse/DST-Assessment-01/tree/main/Data) in the making of this work; stored up in a single place for ease.
